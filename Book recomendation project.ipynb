## popularity based recommendation system

```python
rate_with_name = ratings.merge(books, on='ISBN')
```

```python
rate_with_name.head()
```

```python
num_rating_df = rate_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_rating'},inplace=True)
```

```python
num_rating_df
```

```python
avg_rating_df = rate_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
```

```python
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
```

```python
popular_df
```

```python
popular_df = popular_df[popular_df['num_rating'] >= 250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_rating','avg_rating']]
```

# collabarative based recommendation system

```python
rate_with_name.head()
```

```python
x = rate_with_name.groupby('User-ID').count()['Book-Rating'] > 250
rated_users=x[x].index

```

```python
filterd_rating = rate_with_name[rate_with_name['User-ID'].isin(rated_users)]
```

```python
y = filterd_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index
```

```python
final_ratings = filterd_rating[filterd_rating['Book-Title'].isin(famous_books)]
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
s_score = cosine_similarity(pt)
```

```python
s_score.shape
```

```python
s_score[1]
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_item = sorted(list(enumerate(s_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    for i in similar_item:
        print(pt.index[i[0]])
```

```python
recommend('1984')
```

```python
recommend('The Rainmaker')
```
